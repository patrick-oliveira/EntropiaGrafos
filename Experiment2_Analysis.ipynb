```python
from pathlib import Path
import matplotlib.pyplot as plt
import pickle
from itertools import product

statistics_path_str = "Experiments/Experiment 2/"
statistics_path = Path(statistics_path_str)
```

```python
statistics = pickle.load(open(statistics_path / "-1_0.pickle", "rb"))
```

```python
statistics
```

```python
gammas = [-5, -1, 0, 1, 5]
kappas = [0, 1, 5, 10, 15, 20, 25]
T = 150
Xs = range(T)

fig, axs = plt.subplots(nrows = len(gammas), figsize = (8, 35))
fig.subplots_adjust(hspace = 0.2)

for i in range(len(gammas)):
    gamma = gammas[i]
    for kappa in kappas:
        statistics = pickle.load(open(statistics_path / f"{gamma}_{kappa}.pickle", "rb"))
    
        axs[i].set_title(f"{gamma}", fontsize = 12)
        axs[i].plot(Xs, statistics[f"{gamma}, {kappa}"]['Entropy'], label = kappa)
    axs[i].legend()
```
