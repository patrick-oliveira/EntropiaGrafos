```python
import pickle
from pathlib import Path
from itertools import product
import matplotlib.pyplot as plt
import numpy as np
```

```python
results = pickle.load(open("simulation_results.pickle", "rb"))
```

```python
parameters = {'network_size': [1500], 'memory_size': [100], 'prefferential_att': [2], 'code_length': [5], 'kappa': [0, 5, 10, 15], 'gamma': [-7, -5, -3, -1, 0, 1, 3, 5, 7], 'lambda': [0], 'alpha': [0], 'omega': [0], 'T': 150, 'num_repetitions': 5, 'seed': 42, 'path_str': Path('experiments/experiment_2')}
```

```python
params_cartesian_product = product(parameters['network_size'], parameters['memory_size'], 
                                   parameters['prefferential_att'], parameters['code_length'], 
                                   parameters['kappa'], parameters['gamma'], parameters['lambda'], 
                                   parameters['alpha'], parameters['omega'])

params_cartesian_product = list(params_cartesian_product)

#kappa idx 4
#gamma idx 5
```

```python
fig, axs = plt.subplots(ncols = 4, figsize = (30, 5))

kappas = [0, 5, 10, 15]
gammas = [-7, -5, -3, -1, 0, 1, 3, 5, 7]
X = np.arange(150)

for k in range(len(kappas)):
    axs[k].set_title(f"Kappa = {kappas[k]}")
    for g in range(len(gammas)):
        params = (1500, 100, 2, 5, kappas[k], gammas[g], 0, 0, 0)
        entropy = results[params][0]['Entropy']
        
        axs[k].plot(X, entropy, label = f"g = {gammas[g]}")
    
    axs[k].legend()
```
