```python
import matplotlib.pyplot as plt
import pickle
from itertools import product
```



Aparentemente, aumentar o tamanho do grafo resulta em uma curva mais suave. A dinâmica do sistema é a mesma em todos os casos. O tamanho da memória não parece influenciar demais o resultado da curva.

Para o paper, pode ser razoável escolher um modelo de 2000 vértices e memória de 150 códigos.


```python
Ns = [500, 1000, 1500, 2000, 2500, 3000]
memory_sizes = [50, 100, 150, 200, 250, 300]
kappas = [0, 5, 10]
T = 150
Xs = range(T)
index_product = product(range(6), range(6))

fig, axs = plt.subplots(nrows = len(Ns), ncols = len(memory_sizes), figsize = (22, 18))
fig.subplots_adjust(hspace = 0.5, wspace = 0.5)

fig.suptitle("Curvas de Entropia em Função de N, mu e kappa", fontsize = 25)

for i, j in index_product:
    N = Ns[i]
    memory_size = memory_sizes[j]
    statistics = pickle.load(open(f"{N}_{memory_size}.pickle", "rb"))[f"{N}, {memory_size}"]
    
    axs[i][j].set_title(f"{N}, {memory_size}", fontsize = 12)
    for kappa in kappas:
        axs[i][j].plot(Xs, statistics[kappa]['Entropy'])
```
