```python
import matplotlib.pyplot as plt
import pickle
from itertools import product
```

```python
kappas = [5, 10, 15, 20]
T = 250
Xs = range(T)
lambdas = [0, 1, 5, 10, 15, 20, 25]

fig, axs = plt.subplots(nrows = len(kappas), figsize = (8, 20))
fig.subplots_adjust(hspace = 0.2, wspace = 0.2)
fig.suptitle("Curvas de Entropia em Funo de Kappa e Lambda", fontsize = 25)

for i in range(len(kappas)):
    for lambd in lambdas:
        kappa = kappas[i]
        statistics = pickle.load(open(f"{kappa}_{lambd}.pickle", "rb"))

        axs[i].set_title(f"Kappa = {kappa}", fontsize = 12)
        axs[i].plot(Xs, statistics['Entropy'], label = f"Lambda = {lambd}")
        axs[i].legend()
```
