```python
import matplotlib.pyplot as plt
import pickle
from itertools import product
```

```python
gammas = [-5, -1, 0, 1, 5]
kappas = [0, 1, 5, 10, 15, 20, 25]
T = 150
Xs = range(T)

fig, axs = plt.subplots(nrows = len(gammas), figsize = (8, 35))
fig.subplots_adjust(hspace = 0.2)
fig.suptitle("Curvas de Entropia em Funo de Gamma e Kappa", fontsize = 15)

for i in range(len(gammas)):
    gamma = gammas[i]
    for kappa in kappas:
        statistics = pickle.load(open(f"{gamma}_{kappa}.pickle", "rb"))
    
        axs[i].set_title(f"Gamma = {gamma}", fontsize = 12)
        axs[i].plot(Xs, statistics[f"{gamma}, {kappa}"]['Entropy'], label = f"K = {kappa}")
    axs[i].legend()
```
