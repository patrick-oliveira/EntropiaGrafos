```python

def string_to_binary(x: str):
    return np.asarray(list(x)).astype(int)

def random_selection(distribution, x):
    '''
    Input:
        distribution: A probability distribution over a list of binary codes.
        
    Select randomly a binary code from a given distribution.
    '''
    cumulative_probability = 0
    print(x)
    for code in distribution.keys():
        cumulative_probability += distribution[code]
        print(code, distribution[code], cumulative_probability)
        if cumulative_probability >= x:
            return string_to_binary(code)
        
        
def _random_selection(distribution, x):
    '''
    Input:
        distribution: A probability distribution over a list of binary codes.
        
    Select randomly a binary code from a given distribution.
    '''
    distribution = list(distribution.values())
    cumulative_probability = 0
    print(x)
    for code in A.keys():
        cumulative_probability += distribution[code]
        print(code, distribution[code], cumulative_probability)
        if cumulative_probability >= x:
            return A[code]
```

```python
def random_selection(distribution):
    '''
    Input:
        distribution: A probability distribution over a list of binary codes.
        
    Select randomly a binary code from a given distribution.
    '''
    distribution_ = list(distribution.values())
    x = np.random.uniform()
    cumulative_probability = 0
    for code in A.keys():
        cumulative_probability += distribution_[code]
        if cumulative_probability >= x:
            return A[code], code
```
