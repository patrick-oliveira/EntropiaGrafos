```python
%load_ext autoreload
%autoreload 2

from pathlib import Path
import matplotlib.pyplot as plt
import pickle
from itertools import product

statistics_path_str = "Experiments/Experiment 3/"
statistics_path = Path(statistics_path_str)
```

```python
kappas = [5, 10, 15, 20]
T = 250
Xs = range(T)
lambdas = [0, 1, 5, 10, 15, 20, 25]

fig, axs = plt.subplots(nrows = len(kappas), figsize = (8, 20))
fig.subplots_adjust(hspace = 0.2, wspace = 0.2)

for i in range(len(kappas)):
    for lambd in lambdas:
        kappa = kappas[i]
        statistics = pickle.load(open(statistics_path / f"{kappa}_{lambd}.pickle", "rb"))

        axs[i].set_title(f"{kappa}", fontsize = 12)
        axs[i].plot(Xs, statistics['Proximity'], label = lambd)
        axs[i].legend()
```
